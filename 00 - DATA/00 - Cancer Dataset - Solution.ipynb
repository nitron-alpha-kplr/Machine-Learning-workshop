### The dataset :

In the dataset there are 569 samples and these are digitalized images.   
There are 30 features that describe each sample and each sample is labeled malignant or benign.  

So in machine learning we feed the data into the algorithm telling iy : here is a sample, providing a label telling wether the tumor is a malignant tumor, or a benign tumor.  
And this is how the algorithm learns. After we train the algorithm wa can test it against new data where we only specify the features to ask it whether the sample is malignant or benign.  

The dataset is already cleaned and pre-processed.  

Important : First split the data into training and test 
(it could not be consistent to train and test the classifier on the same set of data, as we would have 100% accuracy).
